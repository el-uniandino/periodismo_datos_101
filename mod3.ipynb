# Los promedios...

```python
# example 4
count = np.array([12,14,11,11,10,11,9,7.5,6.2,6.8,0.1,0.05,0.1,1,1.1,0.7])*100000
income = np.array([0,50,100,150,200,250,300,350,400,450,500,550,600,650,700,750])
mean = round(sum(income*count)/sum(count))
median = 175
mode = 50
a4_dims = (11, 8)
fig, ax = plt.subplots(figsize=a4_dims)
ax.vlines(np.array([mean]),alpha=1, color='orange',ymin=0,ymax=15*100000,linestyle='dashed',label='Promedio')
ax.vlines(np.array([median]),alpha=1, color='black',ymin=0,ymax=15*100000,linestyle='dashed',label='Mediana')
ax.vlines(np.array([mode]),alpha=1, color='red',ymin=0,ymax=15*100000,linestyle='dashed',label='Moda')
ax.bar(income,
       count,
       width=50,
       color='teal')
ax.tick_params(axis='x',
               labelrotation=0)
ax.set_xlabel(r'Ingresos en millones de pesos',fontsize=12)
ax.set_ylabel(r'Count',fontsize=12)
ax.set_title(r'Distribución de ingresos anuales',fontsize=15)
ax.legend()
print(f"Mediana:{median}, Promedio:{mean}")
```
